```python
from sklearn.preprocessing import StandardScaler

features = ['CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT','MEDV']

x = df.loc[:,features].values
y = df.loc[:,['CAT. MEDV']].values

x = StandardScaler().fit_transform(x)
```

```python
# ì¶”ê°€ì ìœ¼ë¡œ,
# boston.csvì— ì£¼ì–´ì§„ ê° ë…ë¦½ë³€ìˆ˜ë“¤ ì¤‘ì—
# ì¢…ì†ë³€ìˆ˜ì— ì˜í–¥ì„ ë¯¸ì¹˜ëŠ” ë³€ìˆ˜ë“¤ì´ ìžˆëŠ”ì§€ í™•ì¸í•˜ê¸° ìœ„í•´ PCAê¸°ë²•ì„ ì‚¬ìš©í–ˆë‹¤.

from sklearn.decomposition import PCA

pca = PCA(n_components=2)

principalComponents = pca.fit_transform(x)

principalDf = pd.DataFrame(data = principalComponents, columns=['principal component 1','principal component 2'])
```

```python
finalDf = pd.concat([principalDf, df[['CAT. MEDV']]], axis = 1)
finalDf.rename(columns = {"principal component 1" : "component1"}, inplace = True)
finalDf.rename(columns = {'principal component 2' : 'component2'}, inplace = True)
```
