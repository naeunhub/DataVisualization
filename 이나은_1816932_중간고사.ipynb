```python
from sklearn.preprocessing import StandardScaler

features = ['CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT','MEDV']

x = df.loc[:,features].values
y = df.loc[:,['CAT. MEDV']].values

x = StandardScaler().fit_transform(x)
```

```python
# 추가적으로,
# boston.csv에 주어진 각 독립변수들 중에
# 종속변수에 영향을 미치는 변수들이 있는지 확인하기 위해 PCA기법을 사용했다.

from sklearn.decomposition import PCA

pca = PCA(n_components=2)

principalComponents = pca.fit_transform(x)

principalDf = pd.DataFrame(data = principalComponents, columns=['principal component 1','principal component 2'])
```

```python
finalDf = pd.concat([principalDf, df[['CAT. MEDV']]], axis = 1)
finalDf.rename(columns = {"principal component 1" : "component1"}, inplace = True)
finalDf.rename(columns = {'principal component 2' : 'component2'}, inplace = True)
```

```python
finalDf
```

```python
groups = finalDf.groupby('CAT. MEDV')

fig, ax = plt.subplots()

for name, group in groups:
    ax.plot(group.component1, 
            group.component2, 
            marker='o', linestyle='', label=name)

ax.legend(fontsize=12, loc='upper left')

plt.xlabel('component1', fontsize=14)

plt.ylabel('component2', fontsize=14)

plt.show()

# PCA 기법을 통해 14개의 독립변수에서
# 종속변수에 가장 영향을 많이 미치는 2개의 성분을 추출하여 그래프를 그림으로써
# 독립변수와 종속변수 간의 관계가 있음을 시각적으로 표현하였다.
```
